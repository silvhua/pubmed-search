# Iteration 2: Replace `None` with `''` in JSON

```python
dictionary_list = load_json(filename, filepath)
dictionary_list
```

```python
def replace_none_with_empty(input_list):
    return [{key: '' if value is None else value for key, value in dictionary.items()} for dictionary in input_list]


updated_list = replace_none_with_empty(dictionary_list)
print(updated_list)
```
